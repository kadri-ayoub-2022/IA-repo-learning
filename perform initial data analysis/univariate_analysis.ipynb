```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("operations.csv",parse_dates=[0])
data.head()

# Columns

# transaction_date → Date of the transaction (e.g., 2023-04-03).

# label → Text label/description of the operation (e.g., PAYMENT XX MTA XX).

# amount → Transaction amount:

# Negative = debit (money spent/withdrawn).

# Positive = credit (money received).

# balance_bef_trn → Account balance before the transaction.

# categ → Category of spending (e.g., TRANSPORT, OTHER).

# type → Type of transaction (e.g., PAYMENT, TRANSFER, DONATION).

# expense_slice → Classifies spending size:

# small → small transaction.

# (not an expense) → not a debit (credits like transfers, salary, refunds).

# debcr → Whether it’s a debit (money out) or credit (money in).

# year, month, day → Date broken into components (useful for time-based analysis).

# day_week → Day of the week (e.g., Monday).

# day_week_num → Numeric representation of day of week (Monday = 1, … Sunday = 7).

# weekend → Boolean → True if the transaction happened on weekend, else False.

# quart_month → Quarter of the month:

# 1 → beginning of month (days 1–7).

# 2 → second week, etc.

# Helps analyze spending habits across the month.
```

```python
import numpy as np

expenses = data[data['amount'] < 0]
exp = -expenses['amount'].values
n = len(exp)    
print(n)
lorenz = np.cumsum(np.sort(exp)) / exp.sum()
lorenz = np.append([0],lorenz) # The Lorenz Curve begins at 0

plt.axes().axis('equal')
xaxis = np.linspace(0-1/n,1+1/n,n+1) # There is 1 segment (of size n) for each individual, plus 1 segment at y=0. The first segment starts at 0-1/n and the last one finishes at 1+1/n
plt.plot(xaxis,lorenz,drawstyle='steps-post')
plt.show()
```

```python
AUC = (lorenz.sum() -lorenz[-1]/2 -lorenz[0]/2)/n # area under the Lorenz Curve. The first segment (lorenz[0]) is halfly below O, so we divide it by 2. We do the same for the mast segment lorenz[-1]
S = 0.5 - AUC # area between 1st bisector and the Lorenz Curve
gini = 2*S
gini
```
